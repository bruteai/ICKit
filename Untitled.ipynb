```python
import matplotlib.pyplot as plt
from PIL import Image
import os
import random

```

```python
folder = 'data/'
images=random.sample(os.listdir(folder), 9)

for i,filename in enumerate(images):
    plt.subplot(330 + 1 + i)
    image = Image.open(os.path.join(folder,filename))
    plt.title(filename.split('.')[0],fontdict={'fontsize': 10,'verticalalignment': 'baseline'})
    plt.imshow(image)
    plt.axis('off')
    print(image.size)
plt.show()
```
